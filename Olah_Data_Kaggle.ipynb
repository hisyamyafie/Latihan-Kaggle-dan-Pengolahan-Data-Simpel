```python
import pandas as pd
import numpy as np
```

```python
path1 = "D:/KULIAH/Semester 5/AI/customers.csv"
path2 = "D:/KULIAH/Semester 5/AI/transactions_1k.csv"
path3 = "D:/KULIAH/Semester 5/AI/gasstations.csv"
path4 = "D:/KULIAH/Semester 5/AI/products.csv"
data_cus = pd.read_csv(path1,'r', delimiter=',')
data_tra = pd.read_csv(path2,'r', delimiter=',')
data_gas = pd.read_csv(path3,'r', delimiter=',')
data_pro = pd.read_csv(path4,'r', delimiter=',')
```

```python
data_cus = data_cus.drop(columns = "Segment")
data_cus
```

```python
data_gas.head()
```

```python
data_pro.head()
```

```python
df_merge = (data_cus.merge(data_tra).merge(data_pro).merge(data_gas))
```

```python
df_merge
```

Data Customer 

```python
df_custable = df_merge.loc[:,"CustomerID":"Price"]
df_custable = df_custable.drop(columns  = ["CardID", "GasStationID","ProductID"])
df_custable
```

Data Gas stassion

```python
df_gastable = df_merge.loc[:,["GasStationID", "Country", "Segment","Time", "Date", "Amount","Price"]]
df_gastable
```

Data Products

```python
df_protable = df_merge.loc[:,["ProductID", "Description", "Time", "Date", "Amount","Price"]]
df_protable
```

# Jawaban Top Customers

```python
cus_price = df_custable.groupby("CustomerID").sum()
```

```python
df_top_cus = cus_price.sort_values("Price", ascending = False).head()
df_top_cus
```

# Jawaban Top Gas Stassion

```python
gas_price = df_gastable.groupby("GasStationID").sum()
```

```python
df_top_gas = gas_price.sort_values("Price", ascending = False).head()
df_top_gas
```

# Jawaban Top Products

```python
pro_price = df_protable.groupby("Description").sum()
df_top_pro = pro_price.sort_values("Price", ascending= False).head()
df_top_pro
```
